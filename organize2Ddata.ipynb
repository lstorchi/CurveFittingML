```python
import pandas as pd
import numpy as np

data1 = pd.read_excel('dv1.xlsx')
data2 = pd.read_excel('dv2.xlsx')
data3 = pd.read_excel('dv3.xlsx')

temp1 = list(data1.columns)[1:]
temp2 = list(data2.columns)[1:]
temp3 = list(data3.columns)[1:]

vib1 = data1[list(data1.columns)[0]].values 
vib2 = data2[list(data2.columns)[0]].values
vib3 = data3[list(data3.columns)[0]].values
```

```python
for i, t in enumerate(temp1):
    ys = data1[list(data1.columns)[i+1]].values
    for j, y in enumerate(ys):
        print(int(vib1[j]), int(t), y)
```

```python
for i, t in enumerate(temp2):
    ys = data2[list(data2.columns)[i+1]].values
    for j, y in enumerate(ys):
        print(int(vib2[j]), int(t), y)
```
